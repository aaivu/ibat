```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('bus_stop_times_feature_added_all.csv')
```

```python
df.conditions.describe()
```

```python
df.stop_type.describe()
```

```python
conditions_le = LabelEncoder()
```

```python
df['conditions'] = conditions_le.fit_transform(df['conditions'])
```

```python
dict(zip(conditions_le.classes_, conditions_le.transform(conditions_le.classes_)))
```

```python
stop_type_le = LabelEncoder()
```

```python
df['stop_type'] = stop_type_le.fit_transform(df['stop_type'])
```

```python
dict(zip(stop_type_le.classes_, stop_type_le.transform(stop_type_le.classes_)))
```

```python
df.to_csv('bus_stop_times_feature_encoded_all.csv', index=False)
```
